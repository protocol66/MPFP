```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read in csv into dataframe
val_acc_df = pd.read_csv("./lightning_logs/version_2/run-version_2-tag-val_acc.csv")
val_loss_df = pd.read_csv("./lightning_logs/version_2/run-version_2-tag-val_loss.csv")

train_acc_df = pd.read_csv("./lightning_logs/version_2/run-version_2-tag-train_acc.csv")
train_loss_df = pd.read_csv("./lightning_logs/version_2/run-version_2-tag-train_loss.csv")

dfs ={"train_acc": train_acc_df, "train_loss": train_loss_df, "val_acc": val_acc_df, "val_loss": val_loss_df}
```

```python
fig, ax = plt.subplots()

acc = ax.twinx()

p1, = acc.plot(train_acc_df["Step"], train_acc_df["Value"], label="train_acc", color="lightblue", linestyle="dashed")
P2, = acc.plot(val_acc_df["Step"], val_acc_df["Value"], label="val_acc", color="blue", linestyle="solid")

p3, = ax.plot(train_loss_df["Step"], train_loss_df["Value"], label="train_loss", color="lightgreen", linestyle="dashed")
p4, = ax.plot(val_loss_df["Step"], val_loss_df["Value"], label="val_loss", color="green", linestyle="solid")

# set labels
ax.set_xlabel("Step")
ax.set_ylabel("Loss")
acc.set_ylabel("Accuracy")
plt.title("Training and Validation Loss and Accuracy")

plt.legend(handles=[p1, P2, p3, p4], loc="upper left", bbox_to_anchor=(1.1, 1.0))
```

```python
fig.savefig("plots.png", bbox_inches="tight", dpi=300)
```
